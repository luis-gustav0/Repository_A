# Funcão

```python
def Fraude(df,col):
  Q1 = df[col].quantile(.25)
  Q3 = df[col].quantile(.75)
  IQR = Q3 - Q1
  limite_inferior = Q1 - 1.5 * IQR
  limite_superior = Q3 + 1.5 * IQR

  outliers_index = (df[col] < limite_inferior)|(df[col] > limite_superior)

  return df.loc[outliers_index]
```

```python
Fraude(df,'Valor da transação')
```
