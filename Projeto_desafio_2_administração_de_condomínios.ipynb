```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_json('/content/dados_locacao_imoveis.json')
```

```python
dados = pd.json_normalize(dados['dados_locacao'])
```

```python
dados.head(10)
```

```python
columns_names = dados.columns.to_list()
```

```python
dados = dados.explode(columns_names[1:])
```

```python
dados.reset_index(drop=True, inplace=True)
```

```python
dados.info()
```

```python
dados['valor_aluguel'] = dados['valor_aluguel'].apply(lambda x: x.replace('$','').replace(',','.').replace('reais','').strip()).astype(np.float64)
```

```python
dados['apartamento'] = dados['apartamento'].str.replace('(blocoAP)','')

# Realizar a substituição dos dados na coluna textual
# dados['apartamento'] = dados['apartamento'].str.replace(' \(blocoAP\)', '', regex=True)
```

```python
date_columns = ['datas_combinadas_pagamento', 'datas_de_pagamento']
```

```python
dados[date_columns] = dados[date_columns].apply(pd.to_datetime, format='%d/%m/%Y')
```

```python
dados['atraso'] = (dados['datas_de_pagamento'] - dados['datas_combinadas_pagamento']).dt.days
```

```python
media_atraso = dados.groupby(dados['apartamento'])['atraso'].mean()
media_atraso
```
