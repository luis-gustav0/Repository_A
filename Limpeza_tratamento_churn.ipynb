***Entendendo os dados***

---


**A base de dados contém colunas além do ID dos clientes e o churn:**

Cliente:

*   genero: gênero (masculino e feminino)
*   idoso: informação sobre um cliente ter ou não idade igual ou maior que 65 anos
*   parceiro: se o cliente possui ou não um parceiro ou parceira
*   dependentes: se o cliente possui ou não dependentes
*   tempo_servico: meses de contrato do cliente


Serviço de telefonia:

*   servico_telefone: assinatura de serviço telefônico
*   varias_linhas: assinatura de mais de uma linha de telefone

Serviço de internet:

*   servico_internet: assinatura de um provedor internet
*   seguranca_online: assinatura adicional de segurança online
*   backup_online: assinatura adicional de backup online
*   protecao_dispositivo: assinatura adicional de proteção no dispositivo
*   suporte_tecnico: assinatura adicional de suporte técnico, menos tempo de espera
*   tv_streaming: assinatura de TV a cabo
*   filmes_streaming: assinatura de streaming de filmes

Conta:

*   contrato: tipo de contrato
*   faturamente_eletronico: se o cliente prefere receber online a fatura
*   metodo_pagamento: forma de pagamento
*   cobranca.mensal: total de todos os serviços do cliente por mês
*   cobranca.Total: total gasto pelo cliente

### dados nulos

Existem alguns métodos da biblioteca Pandas que nos auxiliam a identificar quais são dados nulos em nosso DataFrame, como:

***1 - isnull()***: retorna um DataFrame de valores booleanos, no qual True indica o valor nulo e False indica o valor válido.

***2 - notnull()***: retorna um DataFrame de valores booleanos, no qual True indica o valor válido e False indica o valor nulo.

***3 - isna()***: retorna um DataFrame de valores booleanos, com True indicando o valor nulo e False o valor válido (a mesma funcionalidade do método isnull()).

***4 - notna()***: retorna um DataFrame de valores booleanos, em que True indica o valor válido e False indica o valor nulo (a mesma funcionalidade do método notnull()).

# trabalhando com *outliers*

## identificando outliers

{32076CFD-F9E1-48C2-B72D-9876C41ACFD4}.png

```python
dados_sem_nulo.describe()
```

```python
sns.boxplot(x=dados_sem_nulo['cliente.tempo_servico'])
```

```python
Q1 = dados_sem_nulo['cliente.tempo_servico'].quantile(.25)
Q3 = dados_sem_nulo['cliente.tempo_servico'].quantile(.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
```

```python
outliers_index = (dados_sem_nulo['cliente.tempo_servico'] < limite_inferior)|(dados_sem_nulo['cliente.tempo_servico'] > limite_superior)
```

```python
dados_sem_nulo[outliers_index]['cliente.tempo_servico']
```

## subistituindo valores outliers

```python
dados_sem_out = dados_sem_nulo.copy()
```

```python
dados_sem_out.loc[outliers_index, 'cliente.tempo_servico'] = np.ceil(
    dados_sem_out.loc[outliers_index, 'conta.cobranca.Total'] / dados_sem_out.loc[outliers_index, 'conta.cobranca.mensal']
)
```

```python
sns.boxplot(x=dados_sem_out['cliente.tempo_servico'])
```

```python
dados_sem_out[outliers_index][['cliente.tempo_servico','conta.cobranca.mensal','conta.cobranca.Total']]
```

## removendo *outliers*

```python
Q1 = dados_sem_out['cliente.tempo_servico'].quantile(.25)
Q3 = dados_sem_out['cliente.tempo_servico'].quantile(.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers_index = (dados_sem_out['cliente.tempo_servico'] < limite_inferior)|(dados_sem_out['cliente.tempo_servico'] > limite_superior)
outliers_index
```

```python
dados_sem_out[outliers_index]['cliente.tempo_servico']
```

```python
dados_sem_out = dados_sem_out[~outliers_index]
```

```python
sns.boxplot(x=dados_sem_out['cliente.tempo_servico'])
```

```python
dados_sem_out.reset_index(drop=True, inplace=True)
```
